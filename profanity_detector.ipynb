```python
import pandas as pd
import re
```

```python
location1 = "tweets.csv"
location2 = "bad-words.csv"
```

```python
# splits each tweet into tokens
def tokenize(sentence): 
    return re.findall(r'\w+', sentence.lower())
```

```python
# assigns profanity degree to each sentence based on COUNT of tokens that match 
# as a profanity from the bad words or racial slurs files given input(i have taken bad words file as example,
# a file only consisting of racial slurs will work just fine)
def profanity_degree(tokens):
    return sum(1 for i in tokens if i in profane_words)/len(tokens)
```

```python
# for this tweet file I used this particular encoding
df = pd.read_csv(location1, encoding ="ISO-8859-1",index_col=False)
```

```python
with open(location2) as file:
    profane_words = file.readlines()
    profane_words = [profane_word.rstrip() for profane_word in profane_words]
```

```python
df["tokens"] = df.tweets.apply(tokenize)
```

```python
df
```

```python
df["profanity_degree"] = df.tokens.apply(profanity_degree)
```

```python
df[["tweets","profanity_degree"]]
```

```python
# seeing the sentence with most profanities
df[["tweets","profanity_degree"]].sort_values("profanity_degree", ascending=False)
```

```python
# saving output to the final file
df[["tweets","profanity_degree"]].to_csv("tweets_output.csv",index=False)
```
